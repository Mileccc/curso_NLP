```python
import numpy as np 
import matplotlib.pyplot as plt 
import string 
from sklearn.model_selection import train_test_split
```

```python
archivos = ["./07-data/poemas_Benedetti.txt","./07-data/poemas_pablo_neruda.txt"]
```

```python
textos = []
etiquetas = []
```

```python
for etiqueta, indice in enumerate(archivos):
    print(f"{indice} Corresponde a {etiqueta}")
    
    with open(indice, "r", encoding="utf-8") as archivo:
        # Abrir y Recorrer cada archivo
        for line in archivo:
            # Recorrer cada Linea del archivo
            print(line)
            # Eliminar espacios blancos de la derecha y convertir en minúsculas
            line = line.rstrip().lower()
            print(line)
            if line:
                # Eliminar puntuación
                line = line.translate(str.maketrans("","", string.punctuation))
                textos.append(line)
                etiquetas.append(etiqueta)
            print(line)
```

```python
train_text, test_text, Ytrain, Ytest = train_test_split(textos, etiquetas, test_size=0.1, random_state=42)
```

```python
"""
"<unk>" es una convención que se utiliza a menudo en el procesamiento de lenguaje natural para representar palabras desconocidas o fuera del vocabulario. En este caso, se está asignando el índice 0 a esta palabra especial.
"""
indice = 1
indicepalabras = {'<unk>': 0}
```

```python
# Construcción de un Diccionario de Codificación de Palabras a Índice
for texto in train_text:
    tokens = texto.split()
    for token in tokens:
        if token not in indicepalabras:
            indicepalabras[token] = indice
            indice += 1
```

```python
V = len(indicepalabras)
# Creamos una matriz de 1 con las mismas dimensiones que V, con (2692) filas y (2692) columnas
# Las A Representan la Matriz de Transición
A0 = np.ones((V,V)) 
A1 = np.ones((V,V))
# Y tambien vamos a hacer un vector con (2692) dimensiones
# Las pi Representan las probabilidades iniciales de las palabras
pi0 = np.ones(V) 
pi1 = np.ones(V)
```

```python
def compute_counts(text_as_int, A, pi):
    for tokens in text_as_int:
        last_idx = None
        for idx in tokens:
            # Estamos en la primera palabra de la secuencia
            if last_idx is None:
                pi[idx] += 1 # Aumentamos la probabilidad de la palabra en la primera palabra de la secuencia
            else:
                A[last_idx,idx] += 1 # Aumentamos la probabilidad de la palabra en la palabra previa
            last_idx = idx
```

```python
compute_counts([t for t, y in zip(train_text_int, Ytrain) if y == 0], A0, pi0)
compute_counts([t for t, y in zip(train_text_int, Ytrain) if y == 1], A1, pi1)
```

```python
# Normaliza A y pi para que sean matrices de probabilidad válidas
# Convencete de que esto es equivalente a las fórmulas mostradas anteriormente
A0 /= A0.sum(axis=1, keepdims=True)
pi0 /= pi0.sum()

A1 /= A1.sum(axis=1, keepdims=True)
pi1 /= pi1.sum()
```

```python
# Calcula el logaritmo de A y pi, ya que no necesitamos las probabilidades reales
logA0 = np.log(A0)
logpi0 = np.log(pi0)

logA1 = np.log(A1)
logpi1 = np.log(pi1)
```

```python
count0 = sum(y == 0 for y in Ytrain) # Cuenta de etiquetas de clase 0 en Ytrain
count1 = sum(y == 1 for y in Ytrain) # Cuenta de etiquetas de clase 1 en Ytrain
total = len(Ytrain) # Cantidad total de ejemplos de entrenamiento
p0 = count0 / total # Probabilidad a priori de clase 0
p1 = count1 / total # Probabilidad a priori de clase 1
logp0 = np.log(p0) # Logaritmo de la probabilidad a priori de clase 0
logp1 = np.log(p1) # Logaritmo de la probabilidad a priori de clase 1
p0, p1 # Imprime las probabilidades a priori de ambas clases
```

```python
# Construccion de un clasificador
class Classifier:
    def __init__(self, logAs, logpis, logpriors):
        self.logAs = logAs
        self.logpis = logpis
        self.logpriors = logpriors
        self.k = len(logpriors) # número de clases
    
    def _compute_log_likelihood(self, input_, class_):
        logA = self.logAs[class_]
        logpi = self.logpis[class_]
        
        last_idx = None
        logprob = 0
        for idx in input_:
            if last_idx is None:
                # Es el primer token en la secuencia
                logprob += logpi[idx]
            else:
                # Calcula la probabilidad de transición de la palabra amnterior a la actual
                logprob += logA[last_idx, idx]
                
            # Actualiza last_idx para la próxima iteración
            last_idx = idx
        return logprob
    
    def predict(self, inputs):
        predictions = np.zeros(len(inputs))
        for i, input_ in enumerate(inputs):
            # Calcula los logaritmos de las probabilidades posteriores para cada clase
            posteriors = [self._compute_log_likelihood(input_, c) + self.logpriors[c] for c in range(self.k)]
            
            # Elige la clase con la mayor probabilidad posterior como la predicción
            pred = np.argmax(posteriors)
            predictions[i] = pred
        return predictions
```

```python
# Cada arreglo debe estar en orden ya que se asume que las clases indexan estas listas
clf = Classifier([logA0, logA1], [logpi0, logpi1], [logp0, logp1])
```

```python
Ptrain = clf.predict(train_text_int)
print(f"Train acc: {np.mean(Ptrain == Ytrain)}")
```

```python
Ptest = clf.predict(test_text_int)
print(f"Test acc: {np.mean(Ptest == Ytest)}")
```
